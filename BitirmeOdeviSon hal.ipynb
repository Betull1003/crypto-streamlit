```python
import pandas as pd
import numpy as np

df.columns = df.columns.str.strip()

# yüzdeliklerden kurtuldum ve floata çevirdim
percentage_columns = ['1h', '24h', '7d', '30d']
for col in percentage_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('%', '', regex=False)
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')

#fiyat ve para sütunları
def clean_money_column(series):
    return (
        series.astype(str)
        .str.replace(r'[$,\s]', '', regex=True)
        .replace('-', np.nan)
        .replace('', np.nan)
        .astype(float)
    )

money_columns = ['Price', '24h Volume', 'Market Cap']

for col in money_columns:
    df[col] = clean_money_column(df[col])

print(df['Market Cap'].head())
print(df['Market Cap'].dtype)


#Supply sütunlarını sayıya çevirelim
def convert_supply(value):
    value = str(value).strip()
    if 'Million' in value:
        return float(value.replace('Million', '').strip()) * 1e6
    elif 'Billion' in value:
        return float(value.replace('Billion', '').strip()) * 1e9
    elif 'Thousand' in value:
        return float(value.replace('Thousand', '').strip()) * 1e3
    else:
        try:
            return float(value.replace(',', '').strip())
        except ValueError:
            return np.nan # Use np.nan for missing or invalid values

df['Total Supply'] = df['Total Supply'].apply(convert_supply)
df['Circulating Supply'] = df['Circulating Supply'].apply(convert_supply)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import zscore
numeric_cols = [col for col in numeric_cols if col in df.columns]
# Aykırı değer analizi ve baskılama işlemi
def analyze_and_cap_outliers(df, numeric_cols):
    for col in numeric_cols:
        print(f"\n--- {col} Sütunu ---")

        # Boxplot ile aykırı değerleri görselleştir
        plt.boxplot(df[col], vert=False, patch_artist=True)
        plt.title(f"Boxplot of {col}")
        plt.xlabel(col)
        plt.show()

        # IQR yöntemi ile aykırı değer analizi
        Q1 = df[col].quantile(0.01)
        Q3 = df[col].quantile(0.99)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        print(f"IQR: {IQR}, Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        print(f"Aykırı Değer Sayısı: {len(outliers)}")

        # Z-Score yöntemi ile aykırı değer analizi
        z_scores = zscore(df[col])
        outliers_z = df[np.abs(z_scores) > 3]
        print(f"Z-Score ile Aykırı Değer Sayısı: {len(outliers_z)}")

        # Aykırı değerleri sınıra çek
        df[col] = df[col].apply(
            lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x
        )

        # Sınıra çekilmiş veriyi tekrar görselleştir
        plt.boxplot(df[col].dropna(), vert=False, patch_artist=True)
        plt.title(f"Boxplot of {col} (Sınıra Çekilmiş)")
        plt.xlabel(col)
        plt.show()

    return df

df = analyze_and_cap_outliers(df, numeric_cols)
```

```python
# Aykırı değerleri IQR yöntemiyle temizleme

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Temizlenecek sütunlar
columns_to_clean = ['Price', 'Market Cap', 'Circulating Supply']

# Her sütuna IQR temizliği uygula
df_cleaned = df.copy()
for col in columns_to_clean:
    df_cleaned = remove_outliers_iqr(df_cleaned, col)

print("Orijinal veri boyutu:", df.shape)
print("Temizlenmiş veri boyutu:", df_cleaned.shape)

# Her sütun için alt ve üst sınırları belirleyip, dışındaki verileri çıkarıyor.
# Daha temiz ve aykırı değerlerden arındırılmış bir veri seti elde ediliyor.

scaler = RobustScaler()
X_train.loc[:, numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test.loc[:, numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
# Sonsuz ve eksik değerleri temizle
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestRegressor

df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# Özellik Seçimi
selected_features = ["Market Cap", "24h Volume", "Rank"]
X = df[selected_features]
y = df["Price"]

# Güvenli RMSE fonksiyonu
def safe_rmse(y_true, y_pred):
    try:
        rmse = mean_squared_error(y_true, y_pred, squared=False)
        return np.nan if np.isnan(rmse) or np.isinf(rmse) else rmse
    except:
        return np.nan

scoring = {
    "r2": "r2",
    "mse": "neg_mean_squared_error",
    "rmse": make_scorer(safe_rmse)
}

# Modelleme ve Değerlendirme
model = RandomForestRegressor(random_state=42)
results = cross_validate(model, X, y, cv=5, scoring=scoring, return_train_score=True)

# Sonuçlar
print("Ortalama R²:", np.mean(results["test_r2"]))
print("Ortalama MSE:", -np.mean(results["test_mse"]))  # negatif değer pozitif çevrildi
print("Ortalama RMSE:", np.mean(results["test_rmse"]))
```
